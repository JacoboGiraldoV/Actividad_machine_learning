```python

ds_n = ds.loc[:,["TotRmsAbvGrd","GarageArea", "GarageCars","OverallCond","PoolArea","OverallQual","LotArea", "TotalBsmtSF", "1stFlrSF", "2ndFlrSF", "TotRmsAbvGrd", "Fireplaces", "WoodDeckSF", "OpenPorchSF", "EnclosedPorch", "3SsnPorch", "ScreenPorch", "SalePrice"]]
```

```python
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)
dsn_f = clean_dataset(ds_n)
```

```python
dsn_f.SalePrice.describe()
```

# Creamos variable binaria

```python
dsn_f["SalePriceBinary"] = np.where(dsn_f['SalePrice']>214000.000000 , 1, 0)
dsn_f.SalePriceBinary.describe()
```

# Matriz de correlaciÃ³n
Podemos ver que en la matriz de correlacion casi todos los parametros tienden a tener una correlaciÃ³n positiva con la variable binaria

```python
k=1
corr_mat = dsn_f.corr()
f, ax = plt.subplots(figsize=(12, 9))
heatm = sns.heatmap(corr_mat, vmax=0.9, square=True, cmap = 'mako_r');
heatm.text(15, 20, 'Figure {}'.format(k),fontsize=16)
k=k+1
```

```python
corr_feat = dsn_f.corr().nlargest(5,"SalePriceBinary")["SalePriceBinary"].index
c_map = np.corrcoef(dsn_f[corr_feat].values.T)
mask = np.zeros_like(c_map,dtype=bool)
mask[np.triu_indices_from(mask)] = True
plt.figure(figsize=(20,10))
sns.heatmap(c_map,
            annot=True,
            fmt=".3f",
            annot_kws = {"size":12},
            cmap=sns.cubehelix_palette(),
            xticklabels = corr_feat.values,
            yticklabels = corr_feat.values,
            mask=mask)
```

```python
from scipy.stats import norm
```

Podemos ver que la variables binaria (SalePriceBinary) es positivamente simetrica

```python
sns.distplot(dsn_f['SalePriceBinary'], color= 'blue', fit=norm);
plt.title('Figure {}'.format(k), y=-0.5, fontsize = 20)
print("Skewness: %f" % dsn_f['SalePriceBinary'].skew())
print("Kurtosis: %f" % dsn_f['SalePriceBinary'].kurt())
```

```python
fig = plt.figure(constrained_layout=True, figsize=(12,8))
grid = gridspec.GridSpec(ncols=3, nrows=4, figure=fig)

ax1 = fig.add_subplot(grid[0,:])
sns.distplot(dsn_f["SalePriceBinary"],ax=ax1)
ax1.set_title("Histrogram of SalePrice")

ax2 = fig.add_subplot(grid[2:,:2])
stats.probplot(dsn_f["SalePriceBinary"],plot=ax2)
ax2.set_title("QQplot of SalePrice")

ax3 = fig.add_subplot(grid[2:,2])
sns.boxplot(dsn_f["SalePriceBinary"],ax=ax3,orient="v")
ax3.set_title("Boxplot of SalePriceBinary")
plt.show()
```

# Curva ROC

```python
probas = tree.predict_proba(x_train)
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
# Obtener tasas positivas falsas y verdaderas
fpr, tpr, thresholds = roc_curve(y_train, probas[:,0], pos_label=0)
base_fpr, base_tpr, _ = roc_curve(y_train, [1 for _ in range(len(y_train))])
# obtener Ã¡rea debajo de la curva
roc_auc = auc(fpr, tpr)
# PLOT ROC curva
plt.figure(dpi=150)
plt.plot(fpr, tpr, lw=1, color='green', label=f'AUC = {roc_auc:.3f}')
plt.title('ROC Curve for RF classifier')
plt.rcParams['font.size'] = 16
    # Plot both curvas
plt.plot(base_fpr, base_tpr, 'b', label = 'baseline')
plt.legend();
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
y_train = dsn_f['SalePriceBinary'].astype("int")
x_train = dsn_f.loc[:,["TotRmsAbvGrd","GarageArea", "GarageCars","OverallCond","PoolArea","OverallQual","LotArea", "TotalBsmtSF", "1stFlrSF", "2ndFlrSF", "TotRmsAbvGrd", "Fireplaces", "WoodDeckSF", "OpenPorchSF", "EnclosedPorch", "3SsnPorch", "ScreenPorch"]]

Random_F = RandomForestClassifier(n_estimators = 100,max_depth=1, random_state=0)
Random_Fit = Random_F.fit(x_train, y_train)

confusion_matrix(y_train, Random_Fit.predict(x_train))
```

```python
# Tener importancia
from matplotlib import pyplot
importance = Random_Fit.feature_importances_
# Resumir la importancia de la caracterÃ­stica
for i,v in enumerate(importance):
	print('Feature: %0d, Score: %.5f' % (i,v))
# plot importancia de la trama
plt.figure(dpi=130)
plt.rcParams['font.size'] = 10
plt.barh(x_train.columns, Random_Fit.feature_importances_)
```
